## 3.4. Working with Video Captures

```python
cap = cv2.VideoCapture(video_path)

# looping through the whole video frame by frame
for frame_id in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    
    # Capture a frame
    ret, frame = cap.read()
    
    # Transform video from 'BGR to GRAY'
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Show image
    cv2.imshow('Video Player', gray)
    
    # Render 
    if cv2.waitKey(10) & 0XFF == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```

```python
plt.imshow(frame)
```

## 3.5. Writing out Video

```python
# Input path
input_path = os.path.join('videos','classroom.mp4')

# Video Capture
cap = cv2.VideoCapture(input_path)

# Video Properties
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
fps = cap.get(cv2.CAP_PROP_FPS)

# Video Writer
output_path = os.path.join('videos','output_classroom.mp4')

video_writer = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc('P','I','M','1'), fps, (width, height), isColor = False)



# looping through the whole video frame by frame
for frame_id in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    
    # Capture a frame
    ret, frame = cap.read()
    
    # Transform video from 'BGR to GRAY'
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Show image
    cv2.imshow('Video Player', gray)
    
    # Write the frame
    video_writer.write(gray)
    
    # Render 
    if cv2.waitKey(10) & 0XFF == ord('q'):
        break

# Closing Everthing    
cap.release()
cv2.destroyAllWindows()

video_writer.release()
```

# 5. Detect Edges

## 5.1. Import Dependencies

```python
import cv2
import os
from matplotlib import pyplot as plt
```

## 5.2. Load Image

```python
img_path = os.path.join('images', 'bike.jpg')

img = cv2.imread(img_path)
```

```python
img
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Original image shape
img.shape
```

```python
# Resized shape of the image
(int(img.shape[1]/2),int(img.shape[0]/2))
```

```python
# Resizing image (image, (width, height))
img_resize = cv2.resize(img, (int(img.shape[1]/2),int(img.shape[0]/2)))
# Open using openCV
cv2.imshow('Image', img_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 5.3. Edge Detection

```python
# Applying Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Applying GaussianBlur
blur = cv2.GaussianBlur(gray, (5,5), 0)

# Applying Canny
canny = cv2.Canny(blur, threshold1= 180, threshold2= 200)
```

```python
# Resizing image (image, (width, height))
img_resize = cv2.resize(canny, (int(img.shape[1]/2),int(img.shape[0]/2)))
# Open using openCV
cv2.imshow('Image', img_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 5.4. Export Edges

```python
output_path = os.path.join('images')
cv2.imwrite(output_path+'\\edge_bike.jpg', canny)
```
